```python
data=data.dropna()
```

# ** .**

```python
x=data[['id', 'hour', 'hour_bef_temperature', 'hour_bef_precipitation',
       'hour_bef_windspeed', 'hour_bef_humidity', 'hour_bef_visibility',
       'hour_bef_ozone', 'hour_bef_pm10', 'hour_bef_pm2.5']]
y=data['count']
```

# ** .**

   2 . 

```python
from sklearn.model_selection import train_test_split #  sklearn . 
x_train,x_test, y_train, y_test=train_test_split(x,y, test_size=0.3, random_state=42)
```

# **  .**

  4   (1 ) .

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

# **   .** 

$$y={{x-m}\over{\sigma}}$$

         . 

```python
#   . 
mean=x_train.mean(axis=0)
std=x_train.std(axis=0)
```

```python
print(mean)
print(std)
```

# **  .**

```python
x_train-=mean
x_train/=std
x_test-=mean
x_test/=std
```

# **   .**

```python
from tensorflow import keras #   
from tensorflow.keras.models import Sequential #  
from tensorflow.keras.layers import Dense #    

model = Sequential()
model.add(Dense(20, input_dim=10,activation='relu')) #  4  4  , relu  
model.add(Dense(10, activation='relu'))
model.add(Dense(5, activation='relu'))
model.add(Dense(1, activation='linear')) #,  1
opt=keras.optimizers.SGD(learning_rate=0.00002) # ,  
model.compile(optimizer=opt, loss='mse', metrics=['mae'])
```

# ** .** 

epoch , batch_size  .

  100  epoch=5, batch_size=10
100 10   1 .
  5 .

      . 

```python
import time
a=time.time()
hist = model.fit(x_train, y_train, shuffle=True, epochs=20, batch_size=5)
b=time.time()
print(b-a)
```

```python
import matplotlib.pyplot as plt
plt.plot(hist.history['loss'], label='train_loss')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(hist.history['mae'],  label='train_mae')
plt.legend()
plt.show()
```

```python
y_pred=model.predict(x_test)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(y_test,y_pred, alpha=0.3,label='test')
plt.plot(y_test,y_test,color='red')
plt.xlabel('test')
plt.ylabel('prediction')
plt.grid()
plt.legend()
plt.show()
```

# **  **

```python
from sklearn.linear_model import LinearRegression

lin_reg=LinearRegression()
lin_reg.fit(x_train, y_train)

y_pred=lin_reg.predict(x_test)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(y_test,y_pred, alpha=0.3,label='test')
plt.plot(y_test,y_test,color='red')
plt.xlabel('test')
plt.ylabel('prediction')
plt.grid()
plt.legend()
plt.show()
```

# **  **

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg=RandomForestRegressor(n_estimators=10)
forest_reg.fit(x_train, y_train)

y_pred=forest_reg.predict(x_test)
```
